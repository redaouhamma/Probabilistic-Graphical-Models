```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import math
```

```python
data = pd.read_csv('classification_data_HWK1/classificationC.test',sep="\t", names=["a","b","c"])
```

```python
pi=data["c"].mean()
mu_0=(data[data["c"]==0]["a"].mean(),data[data["c"]==0]["b"].mean())
mu_1=(data[data["c"]==1]["a"].mean(),data[data["c"]==1]["b"].mean())
sigma=data[["a","b"]].cov().multiply((data.size-1)/data.size)
```

```python
colors = ['red','green','blue','purple']
data.plot(kind='scatter', x='a', y='b',c='c',cmap=ListedColormap(colors))
A=np.linalg.inv(sigma)
e_1=np.dot(A,np.subtract(mu_0,mu_1))
alpha=-math.log2(pi/(1-pi))+0.5*np.dot(mu_0,np.dot(A,mu_0))-0.5*np.dot(mu_1,np.dot(A,mu_1))
def line(x):
    return -(e_1[0]/e_1[1])*x + alpha/e_1[1]
plt.plot(np.arange(-9,9),line(np.arange(-9,9)), linestyle='solid', color='b')
```

### Logistic Regression

```python
data['d'] = pd.Series(np.ones(data.size))
X = data[["a","b","d"]]

def sigmoid(w):
    return 1/(1+np.exp(-np.dot(X,w)))
def S(w):
    return np.diag(np.multiply(sigmoid(w),sigmoid(-w)))
# def grad(w):
#     grad=np.dot(data["c"]-sigmoid(w),X)
#     return grad
def hess(w):
    hess=np.dot(np.transpose(X),np.dot(S(w),X))
    return np.linalg.pinv(hess)
def z(w):
    return np.dot(X,w)+np.dot(np.linalg.pinv(S(w)),data['c']-sigmoid(w))
```

```python
w_new=np.ones(3)*100
w=np.random.rand(3)
cpt=0
while(np.linalg.norm(w-w_new)>1e-3):
    tmp=w_new
    w_new=np.dot(hess(w),np.dot(np.dot(S(w),z(w)),X))
    w=tmp
    cpt+=1
print(w)
print(w_new)
print(cpt)
```

```python
colors = ['red','green','blue','purple']
data.plot(kind='scatter', x='a', y='b',c='c',cmap=ListedColormap(colors))
print(w[2]/w[1])
def logline(x):
    return -(w[0]/w[1])*x - w[2]/w[1]

plt.plot(np.arange(-9,9),logline(np.arange(-9,9)), linestyle='solid', color='red')
```

### Linear Regression

```python
beta = np.dot(np.dot(X,np.linalg.inv(np.dot(X.T,X))).T,data[["c"]])
s=np.linalg.norm(data[['c']]-np.dot(data[["a","b","d"]],beta))/data.size

colors = ['red','green','blue','purple']
data.plot(kind='scatter', x='a', y='b',c='c',cmap=ListedColormap(colors))

a=-beta[0]/beta[1]
b=-(beta[2]-1+np.sqrt(-s*math.log2(math.pi*s/2)))/beta[1]
def line(x):
    return a*x + b

plt.plot(np.arange(-9,9),line(np.arange(-9,9)), linestyle='solid', color='red')
```
